```python
def build_rhymes(full_text):
    to_print = []
    color_map = {}
    color_idx = 0
    
    for line in full_text:
        for word in line.split():
            word = p.Text(word)
            sylls = word.syllables()
            
            # If word is unknown, add it with no color
            if len(sylls) == 0:
                to_print.append([word.name, None, None])
                continue

            for s in sylls:
                nuc = s.nuclei()[0]
                if nuc not in color_map.keys():
                    color_map[nuc] = color_idx
                    color_idx += 1

                to_print.append([s.token, nuc, color_map[nuc]])
            to_print.append(["SPACE"])
        to_print.append(["BREAK"])
    return to_print
```

```python
full_text = []
with open('figaro.txt', 'r') as f:
    for line in f:
        full_text.append(line.strip("\n"))
        
to_print = build_rhymes(full_text)
```

```python
for elt in to_print:
    if 'SPACE' in elt:
        print(" ", end = '')
    elif 'BREAK' in elt:
        print("\n")
        continue
    else:
        if elt[1]:
            print(elt[0] + bg(elt[2]), end = '')
        else:
            print(elt[0], end = '')
```
